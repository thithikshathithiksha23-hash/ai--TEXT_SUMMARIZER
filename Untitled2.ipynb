```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization")
```

```python
text = """
Artificial Intelligence is transforming the world by enabling machines to learn from data,
make decisions, and perform tasks that typically require human intelligence.
It is widely used in healthcare, finance, education, and many other industries.
AI improves efficiency, reduces human effort, and opens new possibilities.
"""
```

```python
result = summarizer(text, max_length=50, min_length=20, do_sample=False)
print(result[0]['summary_text'])
```
